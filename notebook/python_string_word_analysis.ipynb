# Python — String & Word Analysis Using Dictionaries

This notebook practises core Python string operations by building word-to-value dictionaries.  
Each exercise takes a list of words as input and maps every word to a computed property — such as its length, first/last letter, vowel count, or sorted characters.

**Concepts covered:** `dict`, `str` slicing, `len()`, `upper()`, `lower()`, `sorted()`, loops, and conditionals.

---
## Q1 — Word Length Mapping
Create a dictionary where the key is the word and the value is its length.

**Input:** `["cat", "apple", "dog"]`  
**Expected Output:** `{'cat': 3, 'apple': 5, 'dog': 3}`

```python
words = ["cat", "apple", "dog"]

def word_length_mapping(words):
    word_length_dict = {}

    for word in words:
       word_length_dict[word] = len(word)

    return word_length_dict

result = word_length_mapping(words)
print(result)
```

## Q7 — Word and Vowel Count
Create a dictionary where the key is the word and the value is the count of vowels in that word.

**Input:** `["apple", "dog", "cat"]`  
**Expected Output:** `{'apple': 2, 'dog': 1, 'cat': 1}`

```python
words = ["apple", "dog", "cat"]

def word_and_vowel_count(words):
    vowel_count_dict = {}
    vowels = "aeiou"

    for word in words:
        vowel_count = 0

        for char in word:
            if char in vowels:
                vowel_count += 1

        vowel_count_dict[word] = vowel_count

    return vowel_count_dict

result = word_and_vowel_count(words)
print(result)
```

## Q8 — Word and Consonant Count
Create a dictionary where the key is the word and the value is the number of consonants.

**Input:** `["cat", "apple", "sky"]`  
**Expected Output:** `{'cat': 2, 'apple': 3, 'sky': 3}`

```python
words = ["cat", "apple", "sky"]

def word_and_consonant_count(words):
    consonant_count_dict = {}
    vowels = "aeiou"

    for word in words:
        consonant_count = 0

        for char in word:
            if char not in vowels:
                consonant_count += 1

        consonant_count_dict[word] = consonant_count

    return consonant_count_dict

result = word_and_consonant_count(words)
print(result)
```

## Q13 — Word and Middle Character
Create a dictionary where the key is the word and the value is the middle character.

**Input:** `["cat", "dog", "sun"]`  
**Expected Output:** `{'cat': 'a', 'dog': 'o', 'sun': 'u'}`

```python
words = ["cat", "dog", "sun"]

def word_and_middle_character(words):
    middle_char_dict = {}

    for word in words:
        middle_index = len(word) // 2
        middle_char_dict[word] = word[middle_index]

    return middle_char_dict

result = word_and_middle_character(words)
print(result)
```

## Q14 — Word and Sorted Letters
Create a dictionary where the key is the word and the value is its letters sorted alphabetically.

**Input:** `["dog", "cat", "bat"]`  
**Expected Output:** `{'dog': 'dgo', 'cat': 'act', 'bat': 'abt'}`

```python
words = ["dog", "cat", "bat"]

def word_and_sorted_letters(words):
    sorted_letters_dict = {}

    for word in words:
        sorted_letters_dict[word] = "".join(sorted(word))

    return sorted_letters_dict

result = word_and_sorted_letters(words)
print(result)
```

## Q15 — Word and Length Squared
Create a dictionary where the key is the word and the value is the square of its length.

**Input:** `["hi", "cat", "apple"]`  
**Expected Output:** `{'hi': 4, 'cat': 9, 'apple': 25}`

```python
words = ["hi", "cat", "apple"]

def word_and_length_squared(words):
    length_squared_dict = {}

    for word in words:
        length_squared_dict[word] = len(word) ** 2

    return length_squared_dict

result = word_and_length_squared(words)
print(result)
```
